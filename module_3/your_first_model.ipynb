# Загрузка Pandas и очистка данных

```python
import pandas as pd
```

```python
df = pd.read_csv('main_task_new.csv')
df_kaggle = pd.read_csv('kaggle_task.csv')
df_kaggle['Rating'] = 0
df.head(5)
```

```python
df.info()
```

```python
df['Price Range'].values
df_kaggle['Price Range'].values
```

```python
# transform Price Range to numbers 
price_dict = {'$': 1, '$$ - $$$': 2, '$$$$': 3}
df['Price Range'] = df['Price Range'].map(price_dict)
df['Price Range'].values

df_kaggle['Price Range'] = df_kaggle['Price Range'].map(price_dict)
```

```python
# checking if there are NaNs in the data set
df.isnull().sum(axis=0)
```

```python
# assuming that NaNs in the Number of Reviews means that there were no reviews given, let's fill no values to 0
df['Number of Reviews'].fillna(0, inplace=True)
# let's assume that there were no info on the price
df['Price Range'] = df['Price Range'].fillna(0)
# let's just fill NaNs in the Cuisine Style to [], as it is in the Reviews columns
df['Cuisine Style'].fillna('[]', inplace=True)

df_kaggle['Number of Reviews'].fillna(0, inplace=True)
df_kaggle['Price Range'] = df_kaggle['Price Range'].fillna(0)
df_kaggle['Cuisine Style'].fillna('[]', inplace=True)
```

```python
# check again if there are NaNs left
df.isnull().sum(axis=0)
```

```python
df_kaggle.isnull().sum(axis=0)
```

```python
df_kaggle['Reviews'].fillna(0, inplace=True)
```

```python
# unique number of cities
df.City.nunique()
```

```python
# convert categorical City to dummy variable
df = pd.get_dummies(df, columns=[ 'City',])
df_kaggle = pd.get_dummies(df_kaggle, columns=[ 'City',])
df.head(5)
```

```python
# Cuisine Style preprocessing 

df['Cuisine Style'] = df['Cuisine Style'].apply(
    lambda x: str(x).replace('[', '').replace(']', '').replace("'", '').replace(' ', '').replace('"', ''))

df['Cuisine Style'] = df['Cuisine Style'].apply(lambda x: x.split(','))


df_kaggle['Cuisine Style'] = df_kaggle['Cuisine Style'].apply(
    lambda x: str(x).replace('[', '').replace(']', '').replace("'", '').replace(' ', '').replace('"', ''))

df_kaggle['Cuisine Style'] = df_kaggle['Cuisine Style'].apply(lambda x: x.split(','))
```

```python
# count how many Cuisine Style in one restaurant
df['Number of Cuisine'] = df['Cuisine Style'].apply(lambda x: len(x))

df_kaggle['Number of Cuisine'] = df_kaggle['Cuisine Style'].apply(lambda x: len(x))
```

```python
df.head(5)
```

```python
# Preprocessing Reviews
df['Reviews'] = df['Reviews'].apply(lambda x: str(x).replace('[','').replace(']','').replace(' ','')) 

df_kaggle['Reviews'] = df_kaggle['Reviews'].apply(lambda x: str(x).replace('[','').replace(']','').replace(' ','')) 
```

```python
# find date when a review was left
df['date_review'] = df['Reviews'].str.findall('\d+/\d+/\d+')

df_kaggle['date_review'] = df_kaggle['Reviews'].str.findall('\d+/\d+/\d+')

df['date_review']
```

```python
df['first_review'] = df['date_review'].apply(lambda x: pd.to_datetime(x).min())
df['last_review'] = df['date_review'].apply(lambda x: pd.to_datetime(x).max())


df_kaggle['first_review'] = df_kaggle['date_review'].apply(lambda x: pd.to_datetime(x).min())
df_kaggle['last_review'] = df_kaggle['date_review'].apply(lambda x: pd.to_datetime(x).max())
```

```python
# days betweeb first and last dates of review 
df['days'] = (df['last_review'] - df['first_review']).apply(lambda x: x.days)
df['days'].fillna(df['days'].mean(), inplace=True)


df_kaggle['days'] = (df_kaggle['last_review'] - df_kaggle['first_review']).apply(lambda x: x.days)
df_kaggle['days'].fillna(df_kaggle['days'].mean(), inplace=True)
```

```python
# don't need first and last review columns anymore
df.drop(['first_review','last_review'], axis=1, inplace=True)

df_kaggle.drop(['first_review','last_review'], axis=1, inplace=True)
```

```python
# searching columns in df that are NOT either floar or int
cols_2_drop = [col for col in df.columns if df[str(col)].dtype == object]

cols_2_drop
```

```python
cols_2_drop_kag = [col for col in df_kaggle.columns if df_kaggle[str(col)].dtype == object]

cols_2_drop_kag
```

```python
# drop not needed columns
df = df.drop(cols_2_drop, axis=1)
df.head(5)
```

```python
df_kaggle = df_kaggle.drop(cols_2_drop_kag, axis=1)
df_kaggle.head(5)
```

# Разбиваем датафрейм на части, необходимые для обучения и тестирования модели

```python
# Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
X = df.drop(['Rating'], axis = 1)
y = df['Rating']
```

```python
# Загружаем специальный инструмент для разбивки:
from sklearn.model_selection import train_test_split
```

```python
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" - для тестирования.
# Для тестирования мы будем использовать 25% от исходного датасета.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# Создаём, обучаем и тестируем модель

```python
# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Создаём модель
regr = RandomForestRegressor(n_estimators=100)

# Обучаем модель на тестовом наборе данных
regr.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = regr.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```
